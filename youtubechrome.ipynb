```python
driver = wb.Chrome('/Users/jiyekim/Desktop/bene/chromedriver.exe')
url = ('https://www.youtube.com/results?search_query=%EC%8B%9C%EA%B3%A8%EC%97%AC%ED%96%89')
driver.get(url)
```

```python
body = driver.find_element_by_tag_name('body')
body.send_keys(Keys.PAGE_DOWN)
```

```python
soup = bs(driver.page_source, 'lxml')
```

```python
title =soup.select('a#video-title')
# 영상 제목만 전체 조회
for i in title:
    print(i.text.strip())
```

```python
view = soup.select('span.style-scope.ytd-video-meta-block')
for i in view:
    print(i.text.strip())
```

```python

title_list = []
view_list = []

for i in range(len(title)):
    title_list.append(title[i].text.strip())
    view_list.append(view[i].text.strip())
    
print(title_list)
print(view_list)
```
